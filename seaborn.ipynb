```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# loads the tips dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# distribution plot histogram
# kde is optional and represeents the kernel density estimation
# bins is optional and represents the no. of partitions we may have , too many bins represents the single instances

sns.distplot(tips['total_bill'], kde = True, bins= 10)
```

```python
# creation of joint plots for one col vs other col in a table
# it copmares the x column vs the y column from dataset tip 
# kind is optinal , can have value hex, reg, sactter, kde and many more
sns.jointplot( x = 'total_bill' , y = 'tip' , data = tips, kind = 'scatter')
```

```python
# creation of pair plots
# it crates a plot with x vs y with all the possible pairs of numerical column
# hue is optional, if needs to categoriased based con the categoriacl values of a column
# tips is the dataset dataframe
# moore..
sns.pairplot(tips)
```

```python
# creation of rugplots
#like distplot it take in the single column of a dataframe
sns.rugplot(tips['total_bill'])
```

```python
# kernel denssity estimation of a column
sns.kdeplot(tips['total_bill'])
```

```python
# categorical plots i.e distribution of data on some category
# x vs y where x is categorical and y is a numerical 
sns.barplot(x= 'sex' , y = 'total_bill', data = tips)
```

```python
# categorical plots i.e distribution of data on some category
# x vs y where x is categorical and y is a numerical
# estimator is optional and can choose specific estimators like std dev, mean,median , count or any custom functions
# by deafult estimator is mean
import numpy as np
sns.barplot(x= 'sex' , y = 'total_bill', data = tips, estimator= np.median)
```

```python
# categorical plots i.e distribution of data on some category
# count plot
# it is number of occurences of the categorial values in a column of dataframe, so only one column
# here more makes than females in dataset
sns.countplot(x = 'sex', data = tips)
```

```python
# categorical plots i.e distribution of data on some category
# creation of boxplot
# total_bill given per day of week
sns.boxplot(x = 'day' , y = 'total_bill' , data = tips)
```

```python
# categorical plots i.e distribution of data on some category
# creation of boxplot
# hue is optional rto see further distribution based on the another column within the exixting boxplot
# # total_bill given per day of week of which how many were males and how many females
sns.boxplot(x = 'day' , y = 'total_bill' , data = tips,hue = 'sex')
```

```python
sns.boxplot(x = 'day' , y = 'total_bill' , data = tips,hue = 'sex')
```

```python
# categorical plots i.e distribution of data on some category
# creation f vilolin plot where x is categorical  and y is numerical. 
# it is  data visualization that combines elements of a box plot and a kernel density plot.
# Each "violin" represents a category or group in your data.The width of the violin at any given 
# point represents the estimated probability density of the data at that value.
sns.violinplot(x = 'sex' , y = 'total_bill' , data = tips)
```

```python
# categorical plots i.e distribution of data on some category
# creation f vilolin plot where x is categorical  and y is numerical. 
# it is  data visualization that combines elements of a box plot and a kernel density plot.
# Each "violin" represents a category or group in your data.The width of the violin at any given 
# point represents the estimated probability density of the data at that value. may split the plot
# based on another categorical value in dataframe, with opti0ns like hue and more
sns.violinplot(x = 'day' , y = 'total_bill' , data = tips, hue = 'sex', split = True)
```

```python
# categorical plots i.e distribution of data on some category
# stripplot .it draws a scatter plot wher one varibale is categorical 
sns.stripplot(x='day', y = 'total_bill', data = tips)
              
```

```python
# categorical plots i.e distribution of data on some category
# stripplot .it draws a scatter plot wher one varibale is categorical 
# hue is optional to add more of diffrention based on another categorical category within same striplot
sns.stripplot(x='day', y = 'total_bill', data = tips,jitter = True, hue = 'sex', dodge = True)
```

```python
# categorical plots i.e distribution of data on some category
# swarmpplot .its cobination of stripplot and violin plot
sns.swarmplot(x = 'day', y ='total_bill' , data = tips)
```

```python
# categorical plots i.e distribution of data on some category
# swarmpplot .its cobination of stripplot and violin plot.
# adding a swarm plot on top of a vioilin plot
sns.violinplot(x = 'day', y ='total_bill' , data = tips)
sns.swarmplot(x = 'day', y ='total_bill' , data = tips, color = 'black')
```

```python
# categorical plots i.e distribution of data on some category
# factorplot is the most general form of plots from which every other plot can be obtained using the kind option.
# kind is optioanl to convert the plot inti any of the kind of categorical plots
sns.factorplot(x = 'day', y ='total_bill' , data = tips, kind = 'bar')
```

```python
# matrix plots
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
% matplotlib inline # only in case of jupyter notebook
```

```python
tips.head()
```

```python
flights.head()
```

```python
# matrix plots
# heat maps
# the heat map takes in the dataset in matrix fgorm where the columnns/rows vs columns/rows
# to do tghis we make wither the correlation table or the pivot table
# for tips dataset make a correalation table
tc = tips.corr()
```

```python
tc
```

```python
# creation of heatmap on tips
# annot and cmap are optional 
sns.heatmap(tc, annot = True, cmap = 'coolwarm')
```

```python
# matrix plots
# heat maps
# the heat map takes in the dataset in matrix fgorm where the columnns/rows vs columns/rows
# to do this we make either the correlation table or the pivot table
# for flighst dataset make a pivot table
fp = flights.pivot_table(index = 'month', columns = 'year' , values = 'passengers')
```

```python
fp
```

```python
# matrix plots
# heat maps
# the heat map takes in the dataset in matrix fgorm where the columnns/rows vs columns/rows
# to do this we make either the correlation table or the pivot table
# for flighst dataset make a pivot table , now make a heat map
sns.heatmap(fp)
```

```python
# matrix plots
# heat maps
# the heat map takes in the dataset in matrix fgorm where the columnns/rows vs columns/rows
# to do this we make either the correlation table or the pivot table
# for flighst dataset make a pivot table , now make a heat map
# add optionals as cmap, linecolor, linewidth 
sns.heatmap(fp, cmap = 'magma', linecolor = 'white' , linewidth = 1.5)
```

```python
# matrix plots
# cluster maps
# it clusters the daatset based on teh similarity of the the row and columns
sns.clustermap(fp)
```

```python
# matrix plots
# cluster maps
# it clusters the daatset based on teh similarity of the the row and columns, so that similar groups are close to eachother
# clusters with option standarrd_sacle standarises it to scaling of 1
sns.clustermap(fp, cmap ='coolwarm' , standard_scale=1)
```

```python
# regression plot
# linearmodel plot whre x vs y is ploted from the dataset
sns.lmplot(x='total_bill', y ='tip', data = tips)
```

```python
# regression plot
# linearmodel plot whre x vs y is ploted from the dataset
# hue is optional for separtion of lones absed on two categotrical data
# palette is optional see the matplotlib colorpalette  for avlaible palettes
sns.lmplot(x='total_bill', y ='tip', data = tips, hue = 'sex',markers= ['o','v'],
           palette = 'inferno' , scatter_kws={'s':10})
```

```python
# regression plot
# linearmodel plot whre x vs y is ploted from the dataset
# col.rows is optional to make subplots based on the values of columns 
sns.lmplot(x='total_bill', y ='tip', data = tips, col = 'sex')
```

```python
# regression plot
# linearmodel plot whre x vs y is ploted from the dataset
# col.rows is optional to make subplots based on the values of columns 
sns.lmplot(x='total_bill', y ='tip', data = tips,  row = 'sex', col = 'day')
```

```python
# regression plot
# linearmodel plot whre x vs y is ploted from the dataset
# col.rows is optional to make subplots based on the values of columns 
# change the size and aspect ratio
sns.lmplot(x='total_bill', y ='tip', data = tips, row = 'sex', col= 'day',aspect = 0.6, size = 10)
          
```
